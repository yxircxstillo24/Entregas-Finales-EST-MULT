```python
# Returns
((data.pct_change()+1).cumprod()).plot(figsize=(10, 7))


plt.legend()


plt.title("Returns", fontsize=16)

plt.ylabel('Returns acumulativos', fontsize=14)
plt.xlabel('Año', fontsize=14)


plt.grid(which="major", color='k', linestyle='-.', linewidth=0.5)
plt.show()
```

## Clustering con tickers

Sklearn https://scikit-learn.org/stable/
KMeans en Sklearn: https://scikit-learn.org/stable/modules/generated/sklearn.cluster.KMeans.html

```python
dict_comp = {
 'Amazon': 'AMZN',
 'Apple': 'AAPL',
 'Walgreen': 'WBA',
 'Northrop Grumman': 'NOC',
 'Boeing': 'BA',
 'Lockheed Martin':'LMT',
 'McDonalds': 'MCD',
 'Intel': 'INTC',
 'IBM': 'IBM',
 'Texas Instruments': 'TXN',
 'MasterCard': 'MA',
 'Microsoft': 'MSFT',
 'General Electric': 'GE',
 'American Express': 'AXP',
 'Pepsi': 'PEP',
 'Coca Cola': 'KO',
 'Johnson & Johnson': 'JNJ',
 'Toyota': 'TM',
 'Honda': 'HMC',
 'Exxon': 'XOM',
 'Chevron': 'CVX',
 'Valero Energy': 'VLO',
 'Ford': 'F',
 'Bank of America': 'BAC'
}

comp = sorted(dict_comp.items(), key=lambda x: x[1])
```

```python
type(comp)
```

```python
type(list(dict_comp.values() ) )
```

```python
start_date = '2017-01-01'
end_date = '2019-01-01'


panel = pdr.get_data_yahoo(list(dict_comp.values()), start_date, end_date)
panel
```

Realizaremos un análisis de Cluster sobre los movimientos de las acciones.

```python
import numpy as np
stock_close = panel['Close']
stock_open = panel['Open']
stock_close = np.array(stock_close).T
stock_open = np.array(stock_open).T

row, col = stock_close.shape

movs = np.zeros([row, col])

for i in range(0, row):
 movs[i,:] = np.subtract(stock_close[i,:], stock_open[i,:])
```

```python
movs.shape
```

```python
for i in range(0, len(comp)):
 print('Compañía: {}, Cambio: {}'.format(comp[i][0], sum(movs[i][:])))
```

```python
from sklearn.preprocessing import Normalizer

normalizer = Normalizer()

new = normalizer.fit_transform(movs)
```

```python
from sklearn.pipeline import make_pipeline
from sklearn.cluster import KMeans


kmeans = KMeans(n_clusters=4, max_iter=1000)

pipeline = make_pipeline(normalizer,kmeans)

pipeline.fit(movs)
```

```python
labels = pipeline.predict(movs)

df = pd.DataFrame({'labels': labels, 'compañía': comp})

print(df.sort_values('labels'))
```
